```python
import matplotlib.pyplot as plt
import numpy as np
```

I am interested to get some visualisation diffarential equation. I am using **Vector Slope** method.

# Understanding of `meshgrid`

```python
x = np.linspace(-4, 4, 20)
y = np.linspace(-4, 4, 20)
X, Y = np.meshgrid(x,y)
plt.plot(X,Y, marker = '.', linestyle = 'none')
plt.show()
```

## To Plot:: $\frac{dy}{dx}=\cos(x)$

```python
dy = np.cos(X)
dx = np.ones(dy.shape)

#Plot
plt.quiver(X, Y, dy, dx)
plt.title("Vector plot of y'(x) = sin(x)")
plt.plot(x, np.sin(x))
plt.grid(True)
plt.show()
```

* Make the line unit. Only interested about **Direction**.

```python
dyu = dy/np.sqrt(dx**2 + dy**2)
dxu = dx/np.sqrt(dx**2 + dy**2)
plt.quiver(X, Y, dyu, dxu)
plt.show()
```
